```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,2*np.pi,100)

x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

plt.plot(x,y,c=(1,0.2,0.5),lw=20)  #Pink colour = (red=1, green=0.2, blue=0.5) Width=20


plt.title('Missing you')
plt.axis('equal')
plt.axis('off')

plt.show()
```
